```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from pandas import DataFrame
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
# Importing the dataset
data = pd.read_csv('spamraw.csv')

vec = CountVectorizer()
counts = vec.fit_transform(data['text'].values)

classifier = MultinomialNB()
targets = data['type'].values
classifier.fit(counts,targets)
data

from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
data['type'] = le.fit_transform(data['type'])
data
```

```python
examples = ['Free Vigra Now',"varify your email ID"]
example_counts = vec.transform(examples)
predictions = classifier.predict(example_counts)
predictions
```

```python
# Splitting the dataset into the Training set and Test set
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data['text'], data['type'], test_size = 0.25, random_state = 0)
examp = vec.transform(X_test)
y_pred = classifier.predict(examp)
y_pred = le.fit_transform(y_pred)
y_pred
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_pred)
cm
```
